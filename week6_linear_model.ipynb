```python
from sklearn import linear_model, model_selection, datasets, neighbors
```

```python
data = datasets.load_digits()
X, y = data.data, data.target
```

```python
for model, modelname in [
    [neighbors.KNeighborsClassifier(), 'knn'],
    [linear_model.LogisticRegressionCV(), 'logistic'],
    ]:
    print(modelname) 
    print(model_selection.cross_val_score(model, X, y, cv=5))
```
