```python
%pylab inline
from sklearn import datasets
```

```python
data = datasets.load_iris()
X, y = data.data, data.target
```

```python
from sklearn import model_selection, metrics
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y)
```

```python
from sklearn import neighbors
model = neighbors.KNeighborsClassifier()
model.fit(X_train, y_train)
y_test_predict = model.predict(X_test)
metrics.accuracy_score(y_test, y_test_predict)
```

```python
y_train_predict = model.predict(X_train)
metrics.accuracy_score(y_train, y_train_predict)
```

```python
model_selection.cross_validate(model, X, y, cv = 5)
```

```python
train_scores, test_scores = model_selection.validation_curve(model, X, y, 'n_neighbors', range(1,51), cv = 5)
```
